```python
from pyspark.sql.functions import col, to_timestamp

bronze_df = spark.table("workspace.default.bronze_orders")

silver_df = (
    bronze_df
    .dropDuplicates(["order_id"])
    .dropna()
    .withColumn("quantity", col("quantity").cast("int"))
    .withColumn("price", col("price").cast("double"))
    .withColumn(
        "order_timestamp",
        to_timestamp("order_timestamp")
    )
)
```

```python
from pyspark.sql.functions import *

bronze_df = spark.table("workspace.default.bronze_orders")

display(bronze_df)
```

```python
from pyspark.sql.functions import *

silver_df = (
    bronze_df

    # Remove duplicates
    .dropDuplicates(["order_id"])

    # Remove nulls
    .dropna()

    # Correct data types
    .withColumn("quantity", col("quantity").cast("int"))
    .withColumn("price", col("price").cast("double"))
    .withColumn("order_timestamp", to_timestamp("order_timestamp"))

    # Standardize text
    .withColumn("product", initcap(col("product")))
    .withColumn("payment_method", initcap(col("payment_method")))
    .withColumn("city", initcap(col("city")))

    # Business rule
    .filter(col("quantity") > 0)
    .filter(col("price") > 0)

    # Derived column
    .withColumn("total_amount", col("quantity") * col("price"))

    # Date dimensions
    .withColumn("order_date", to_date(col("order_timestamp")))
    .withColumn("order_year", year(col("order_timestamp")))
    .withColumn("order_month", month(col("order_timestamp")))
    .withColumn("order_day", dayofmonth(col("order_timestamp")))
)
```

```python
from pyspark.sql.functions import sum, count

silver_df = spark.table("workspace.default.silver_orders")

gold_product = (
    silver_df
    .groupBy("product")
    .agg(
        sum("total_amount").alias("total_revenue"),
        count("*").alias("total_orders")
    )
    .orderBy("total_revenue", ascending=False)
)

display(gold_product)
```

```python
gold_product.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.default.gold_sales_by_product")
```

```python
from pyspark.sql.functions import sum, count

gold_city = (
    silver_df
    .groupBy("city")
    .agg(
        sum("total_amount").alias("total_revenue"),
        count("*").alias("total_orders")
    )
    .orderBy("total_revenue", ascending=False)
)

display(gold_city)
```

```python
gold_city.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.default.gold_sales_by_city")
```

```python
gold_daily = (
    silver_df
    .groupBy("order_date")
    .agg(
        sum("total_amount").alias("daily_revenue"),
        count("*").alias("total_orders")
    )
    .orderBy("order_date")
)

display(gold_daily)
```

```python
gold_daily.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.default.gold_daily_sales")
```

```python
gold_payment = (
    silver_df
    .groupBy("payment_method")
    .agg(
        count("*").alias("total_orders"),
        sum("total_amount").alias("total_revenue")
    )
    .orderBy("total_revenue", ascending=False)
)

display(gold_payment)
```

```python
gold_payment.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.default.gold_payment_summary")
```
